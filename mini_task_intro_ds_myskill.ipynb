```python
import pandas as pd
import time


from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('diabetes.csv', sep = ',')

data.head()
```

```python
data.info()
```

#### Berdasarkan info diatas, dapat kita lihat bahwa kesemua data jumlahnya sama ya tidak ada data null, untuk selanjutnya mari kita hapus kolom outcome

```python
data_x = data.drop('Outcome', axis=1)
data_x.head()
```

```python
data_y = data['Outcome']
data_y.head()
```

```python
# Model Training

start = time.time()

X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.33)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

end = time.time()

print(f"Model training completed in {end-start}")
```

```python
# Model Evaluating

y_pred = model.predict(X_test)

accuracy = model.score(X_test, y_test)
print(accuracy)

confussion_result = confusion_matrix(y_test, y_pred)
print(confussion_result)
```
